```python
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from PIL import Image, ImageFilter
import torch
import torch.nn as nn
import numpy as np
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
class CouplingLayer(nn.Module):
    def __init__(self, input_size, hidden_size):
        super().__init__()
        self.input_size = input_size
        self.split_size = input_size // 2
        
        self.scale_net = nn.Sequential(
            nn.Linear(self.split_size, hidden_size),
            nn.ReLU(),
            nn.Linear(hidden_size, hidden_size),
            nn.ReLU(),
            nn.Linear(hidden_size, input_size - self.split_size),
            nn.Tanh()
        )
        
        self.translation_net = nn.Sequential(
            nn.Linear(self.split_size, hidden_size),
            nn.ReLU(),
            nn.Linear(hidden_size, hidden_size),
            nn.ReLU(),
            nn.Linear(hidden_size, input_size - self.split_size)
        )
        
    def forward(self, x):
        x1, x2 = torch.split(x, [self.split_size, self.input_size - self.split_size], dim=1)
        s = self.scale_net(x1)
        t = self.translation_net(x1)
        
        y1 = x1
        y2 = x2 * torch.exp(s) + t
        
        return torch.cat([y1, y2], dim=1), torch.sum(s, dim=1)
    
    def inverse(self, y):
        y1, y2 = torch.split(y, [self.split_size, self.input_size - self.split_size], dim=1)
        s = self.scale_net(y1)
        t = self.translation_net(y1)
        
        x1 = y1
        x2 = (y2 - t) * torch.exp(-s)
        
        return torch.cat([x1, x2], dim=1)

class RealNVP(nn.Module):
    def __init__(self, input_size, hidden_size, blocks):
        super().__init__()
        self.input_size = input_size
        
        self.coupling_layers = nn.ModuleList([
            CouplingLayer(input_size, hidden_size) for _ in range(blocks)
        ])
        
        self.rotations = [torch.qr(torch.randn(input_size, input_size))[0] for _ in range(blocks-1)]
        
    def forward(self, x):
        log_det = torch.zeros(x.shape[0])
        
        for i, coupling in enumerate(self.coupling_layers):
            x, s = coupling(x)
            log_det += s
            
            if i < len(self.coupling_layers) - 1:
                x = x @ self.rotations[i]
                
        return x, log_det
    
    def inverse(self, z):
        x = z
        
        for i, coupling in reversed(list(enumerate(self.coupling_layers))):
            if i < len(self.coupling_layers) - 1:
                x = x @ self.rotations[i].t()
            
            x = coupling.inverse(x)
            
        return x
    
    def sample(self, num_samples):
        z = torch.randn(num_samples, self.input_size)
        return self.inverse(z).detach().numpy()

def train_inn(model, train_data, epochs, lr=0.001, batch_size=128):
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    
    for epoch in range(epochs):
        total_loss = 0
        
        for i in range(0, len(train_data), batch_size):
            batch = torch.FloatTensor(train_data[i:i+batch_size])
            
            optimizer.zero_grad()
            
            
            z, log_det = model(batch)
            
            
            loss = 0.5 * torch.sum(z2, dim=1) - log_det
            loss = torch.mean(loss)
            
            loss.backward()
            optimizer.step()
            
            total_loss += loss.item()
            
        if (epoch + 1) % 10 == 0:
            print(f'Epoch {epoch+1}, Loss: {total_loss/len(train_data)*batch_size:.4f}')


def get_two_moons(n_samples):
    X, _ = make_moons(n_samples=n_samples, noise=0.1)
    return X
```

```python
def get_hexagonal_gmm(n_samples):
    
    radius = 2.0  
    std = radius * 0.1 
    
   
    angles = np.linspace(0, 2*np.pi, 7)[:-1]  
    centers = []
    for angle in angles:
        x = radius * np.cos(angle)
        y = radius * np.sin(angle)
        centers.append([x, y])
    centers = np.array(centers)
    
    center_indices = np.random.choice(len(centers), size=n_samples)
    
    samples = []
    for idx in center_indices:
        center = centers[idx]
        sample = np.random.normal(loc=center, scale=std, size=2)
        samples.append(sample)
    
    return np.array(samples)
```

```python
# Compute MMD
def compute_mmd(X, Y, sigma=1.0):
    X = torch.FloatTensor(X)
    Y = torch.FloatTensor(Y)
    
    xx = torch.cdist(X, X)
    yy = torch.cdist(Y, Y)
    xy = torch.cdist(X, Y)
    
    xx_kernel = torch.exp(-xx2 / (2*sigma2))
    yy_kernel = torch.exp(-yy2 / (2*sigma2))
    xy_kernel = torch.exp(-xy2 / (2*sigma2))
    
    mmd = xx_kernel.mean() + yy_kernel.mean() - 2*xy_kernel.mean()
    
    return mmd.item()


def main():
    input_size = 2
    hidden_size = 64
    blocks = 4
    n_samples = 5000
    epochs = 100
    
    
    model = RealNVP(input_size, hidden_size, blocks)
    X_moons = get_two_moons(n_samples)
    train_inn(model, X_moons, epochs)
    
    
    generated_samples = model.sample(n_samples)
    test_samples = get_two_moons(n_samples)
    mmd = compute_mmd(test_samples, generated_samples)
    
    print(f"MMD for two moons: {mmd:.4f}")
    
   
    plt.figure(figsize=(12, 4))
    
    plt.subplot(131)
    plt.scatter(X_moons[:, 0], X_moons[:, 1], alpha=0.5)
    plt.title("Original Two Moons")
    
    plt.subplot(132)
    plt.scatter(generated_samples[:, 0], generated_samples[:, 1], alpha=0.5)
    plt.title("Generated Two Moons")
    
    
    z, _ = model(torch.FloatTensor(X_moons))
    z = z.detach().numpy()
    
    plt.subplot(133)
    plt.scatter(z[:, 0], z[:, 1], alpha=0.5)
    plt.title("Latent Space")
    
    plt.tight_layout()
    plt.show()
    
    
    model = RealNVP(input_size, hidden_size, blocks)
    X_gmm = get_hexagonal_gmm(n_samples)
    train_inn(model, X_gmm, epochs)
    
    generated_samples = model.sample(n_samples)
    test_samples = get_hexagonal_gmm(n_samples)
    mmd = compute_mmd(test_samples, generated_samples)
    
    print(f"MMD for hexagonal GMM: {mmd:.4f}")
    
    
    plt.figure(figsize=(12, 4))
    
    plt.subplot(131)
    plt.scatter(X_gmm[:, 0], X_gmm[:, 1], alpha=0.5)
    plt.title("Original GMM")
    
    plt.subplot(132)
    plt.scatter(generated_samples[:, 0], generated_samples[:, 1], alpha=0.5)
    plt.title("Generated GMM")
    
    z, _ = model(torch.FloatTensor(X_gmm))
    z = z.detach().numpy()
    
    plt.subplot(133)
    plt.scatter(z[:, 0], z[:, 1], alpha=0.5)
    plt.title("Latent Space")
    
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import torch
import torch.nn as nn
import numpy as np
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

class CouplingLayer(nn.Module):
    def __init__(self, input_size, hidden_size, condition_size):
        super().__init__()
        self.input_size = input_size
        self.split_size = input_size // 2
        
        
        self.scale_net = nn.Sequential(
            nn.Linear(self.split_size + condition_size, hidden_size), 
            nn.ReLU(),
            nn.Linear(hidden_size, hidden_size),
            nn.ReLU(),
            nn.Linear(hidden_size, input_size - self.split_size),
            nn.Tanh()
        )
        
        self.translation_net = nn.Sequential(
            nn.Linear(self.split_size + condition_size, hidden_size), 
            nn.ReLU(),
            nn.Linear(hidden_size, hidden_size),
            nn.ReLU(),
            nn.Linear(hidden_size, input_size - self.split_size)
        )
        
    def forward(self, x, condition):
        x1, x2 = torch.split(x, [self.split_size, self.input_size - self.split_size], dim=1)
        
        conditioned_x1 = torch.cat([x1, condition], dim=1)
        
        s = self.scale_net(conditioned_x1)
        t = self.translation_net(conditioned_x1)
        
        y1 = x1
        y2 = x2 * torch.exp(s) + t
        
        return torch.cat([y1, y2], dim=1), torch.sum(s, dim=1)
    
    def inverse(self, y, condition):
        y1, y2 = torch.split(y, [self.split_size, self.input_size - self.split_size], dim=1)
        conditioned_y1 = torch.cat([y1, condition], dim=1)  
        s = self.scale_net(conditioned_y1)
        t = self.translation_net(conditioned_y1)
        
        x1 = y1
        x2 = (y2 - t) * torch.exp(-s)
        
        return torch.cat([x1, x2], dim=1)


class RealNVP(nn.Module):
    def __init__(self, input_size, hidden_size, blocks, condition_size):
        super().__init__()
        self.input_size = input_size
        self.condition_size = condition_size
        
        self.coupling_layers = nn.ModuleList([
            CouplingLayer(input_size, hidden_size, condition_size) for _ in range(blocks)
        ])
        
        self.rotations = [torch.qr(torch.randn(input_size, input_size))[0] for _ in range(blocks-1)]
        
    def forward(self, x, condition):
        log_det = torch.zeros(x.shape[0])
        
        for i, coupling in enumerate(self.coupling_layers):
            x, s = coupling(x, condition)  
            log_det += s
            
            if i < len(self.coupling_layers) - 1:
                x = x @ self.rotations[i]
                
        return x, log_det
    
    def inverse(self, z, condition):
        x = z
        for i, coupling in reversed(list(enumerate(self.coupling_layers))):
            if i < len(self.coupling_layers) - 1:
                x = x @ self.rotations[i].t()
            x = coupling.inverse(x, condition) 
        return x
    

    def sample(self, num_samples, conditions):
        
        z = torch.randn(num_samples * conditions.shape[0], self.input_size)


        samples = []
        for i in range(conditions.shape[0]):
            condition = conditions[i].repeat(num_samples, 1)
            sample = self.inverse(z[i*num_samples:(i+1)*num_samples], condition)
            samples.append(sample)
            
        return torch.cat(samples).detach().numpy()




def train_inn(model, train_data, train_labels, epochs, lr=0.001, batch_size=128):
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    
    for epoch in range(epochs):
        total_loss = 0
        
        for i in range(0, len(train_data), batch_size):
            batch = torch.FloatTensor(train_data[i:i+batch_size])
            
            
            labels_batch = train_labels[i:i + batch_size]
            condition = torch.FloatTensor(np.eye(model.condition_size)[labels_batch]) 


            optimizer.zero_grad()
            z, log_det = model(batch, condition)
            loss = 0.5 * torch.sum(z2, dim=1) - log_det
            loss = torch.mean(loss)
            loss.backward()
            optimizer.step()
            
            total_loss += loss.item()
            
        if (epoch + 1) % 10 == 0:
            print(f'Epoch {epoch+1}, Loss: {total_loss/len(train_data)*batch_size:.4f}')

def get_two_moons(n_samples):
    X, y = make_moons(n_samples=n_samples, noise=0.1)  
    return X, y

def get_hexagonal_gmm(n_samples, n_classes=6): 
    radius = 2.0
    std = radius * 0.1
    angles = np.linspace(0, 2*np.pi, n_classes+1)[:-1]
    centers = np.array([[radius * np.cos(angle), radius * np.sin(angle)] for angle in angles])
    
    center_indices = np.random.choice(len(centers), size=n_samples)
    samples = np.array([np.random.normal(loc=centers[idx], scale=std, size=2) for idx in center_indices])
    return samples, center_indices 

def compute_mmd(X, Y, sigma=1.0):
    X = torch.FloatTensor(X)
    Y = torch.FloatTensor(Y)
    
    xx = torch.cdist(X, X)
    yy = torch.cdist(Y, Y)
    xy = torch.cdist(X, Y)
    
    xx_kernel = torch.exp(-xx2 / (2*sigma2))
    yy_kernel = torch.exp(-yy2 / (2*sigma2))
    xy_kernel = torch.exp(-xy2 / (2*sigma2))
    
    mmd = xx_kernel.mean() + yy_kernel.mean() - 2*xy_kernel.mean()
    
    return mmd.item()


def main():
    input_size = 2
    hidden_size = 64
    blocks = 4
    n_samples = 200
    epochs = 100

    
    X_moons, y_moons = get_two_moons(n_samples)
    n_classes_moons = len(np.unique(y_moons))
    model_moons = RealNVP(input_size, hidden_size, blocks, n_classes_moons)  
    train_inn(model_moons, X_moons, y_moons, epochs) 


    
    X_gmm6, y_gmm6 = get_hexagonal_gmm(n_samples)
    n_classes_gmm6 = len(np.unique(y_gmm6))
    model_gmm6 = RealNVP(input_size, hidden_size, blocks, n_classes_gmm6)
    train_inn(model_gmm6, X_gmm6, y_gmm6, epochs)



   
    X_gmm2, y_gmm2 = get_hexagonal_gmm(n_samples)
    y_gmm2 = np.where(y_gmm2 < 2, 0, 1)  
    n_classes_gmm2 = len(np.unique(y_gmm2))
    model_gmm2 = RealNVP(input_size, hidden_size, blocks, n_classes_gmm2)
    train_inn(model_gmm2, X_gmm2, y_gmm2, epochs)

    def evaluate_and_visualize(model, X_test, y_test, title, n_classes):
        generated_samples = []
        for label in range(n_classes):
            condition = torch.FloatTensor(np.eye(n_classes)[label]).unsqueeze(0) 
            generated_samples.append(model.sample(n_samples // n_classes, condition.repeat(n_samples // n_classes, 1)))

        generated_samples = np.concatenate(generated_samples)

        
        mmd_marginal = compute_mmd(X_test, generated_samples)
        print(f"MMD (Marginal) for {title}: {mmd_marginal:.4f}")

        
        plt.figure(figsize=(15, 5))
        for label in range(n_classes):
            plt.subplot(1, n_classes + 1, label + 1)
            X_test_label = X_test[y_test == label]
            gen_samples_label = generated_samples[label * (n_samples // n_classes):(label + 1) * (n_samples // n_classes)]
            mmd_conditional = compute_mmd(X_test_label, gen_samples_label)

            plt.scatter(X_test_label[:, 0], X_test_label[:, 1], alpha=0.5, label="Real")
            plt.scatter(gen_samples_label[:, 0], gen_samples_label[:, 1], alpha=0.5, label="Generated")
            plt.title(f"{title} - Label {label} (MMD: {mmd_conditional:.4f})")
            plt.legend()

        # Latent space visualization (using all data)
        z, _ = model(torch.FloatTensor(X_test), torch.FloatTensor(np.eye(n_classes)[y_test]))
        z = z.detach().numpy()
        plt.subplot(1, n_classes + 1, n_classes + 1)
        plt.scatter(z[:, 0], z[:, 1], alpha=0.5, c=y_test) 
        plt.title(f"{title} - Latent Space")


        plt.tight_layout()
        plt.show()


    X_moons_test, y_moons_test = get_two_moons(n_samples)
    evaluate_and_visualize(model_moons, X_moons_test, y_moons_test, "Two Moons", n_classes_moons)


    X_gmm6_test, y_gmm6_test = get_hexagonal_gmm(n_samples)
    evaluate_and_visualize(model_gmm6, X_gmm6_test, y_gmm6_test, "Hexagonal GMM (6 classes)", n_classes_gmm6)

    X_gmm2_test, y_gmm2_test = get_hexagonal_gmm(n_samples)
    y_gmm2_test = np.where(y_gmm2_test < 2, 0, 1)
    evaluate_and_visualize(model_gmm2, X_gmm2_test, y_gmm2_test, "Hexagonal GMM (2 classes)", n_classes_gmm2)

if __name__ == "__main__":
    main()
```

```python
class DownsampleTransform:
    def __init__(self, target_shape, algorithm=Image.Resampling.LANCZOS):
         self.width, self.height = target_shape
         self.algorithm = algorithm
    def __call__(self, img):
         img = img.resize((self.width+2, self.height+2), self.algorithm)
         img = img.crop((1, 1, self.width+1, self.height+1))
         return img
```

```python
class CouplingLayer(nn.Module):
    def __init__(self, input_size, hidden_size, condition_size):
        super().__init__()
        self.input_size = input_size
        self.split_size = input_size // 2
        
        self.scale_net = nn.Sequential(
            nn.Linear(self.split_size + condition_size, hidden_size), 
            nn.ReLU(),
            nn.Linear(hidden_size, hidden_size),
            nn.ReLU(),
            nn.Linear(hidden_size, input_size - self.split_size),
            nn.Tanh()
        )
        
        self.translation_net = nn.Sequential(
            nn.Linear(self.split_size + condition_size, hidden_size), 
            nn.ReLU(),
            nn.Linear(hidden_size, hidden_size),
            nn.ReLU(),
            nn.Linear(hidden_size, input_size - self.split_size)
        )
        
    def forward(self, x, condition):
        x1, x2 = torch.split(x, [self.split_size, self.input_size - self.split_size], dim=1)
        
        # Concatenate condition to x1
        conditioned_x1 = torch.cat([x1, condition], dim=1)
        
        s = self.scale_net(conditioned_x1)
        t = self.translation_net(conditioned_x1)
        
        y1 = x1
        y2 = x2 * torch.exp(s) + t
        
        return torch.cat([y1, y2], dim=1), torch.sum(s, dim=1)
    
    def inverse(self, y, condition):
        y1, y2 = torch.split(y, [self.split_size, self.input_size - self.split_size], dim=1)
        conditioned_y1 = torch.cat([y1, condition], dim=1)  # Condition y1
        s = self.scale_net(conditioned_y1)
        t = self.translation_net(conditioned_y1)
        
        x1 = y1
        x2 = (y2 - t) * torch.exp(-s)
        
        return torch.cat([x1, x2], dim=1)
```

```python
class RealNVP(nn.Module):
    def __init__(self, input_size, hidden_size, blocks, condition_size):
        super().__init__()
        self.input_size = input_size
        self.condition_size = condition_size
        
        self.coupling_layers = nn.ModuleList([
            CouplingLayer(input_size, hidden_size, condition_size) for _ in range(blocks)
        ])
        
        self.rotations = [torch.qr(torch.randn(input_size, input_size))[0] for _ in range(blocks-1)]
        
    def forward(self, x, condition):
        log_det = torch.zeros(x.shape[0])
        
        for i, coupling in enumerate(self.coupling_layers):
            x, s = coupling(x, condition)  
            log_det += s
            
            if i < len(self.coupling_layers) - 1:
                x = x @ self.rotations[i]
                
        return x, log_det
    
    def inverse(self, z, condition):
        x = z
        for i, coupling in reversed(list(enumerate(self.coupling_layers))):
            if i < len(self.coupling_layers) - 1:
                x = x @ self.rotations[i].t()
            x = coupling.inverse(x, condition) 
        return x
    

    def sample(self, num_samples, conditions):
        z = torch.randn(num_samples * conditions.shape[0], self.input_size)


        samples = []
        for i in range(conditions.shape[0]):
            condition = conditions[i].repeat(num_samples, 1)
            sample = self.inverse(z[i*num_samples:(i+1)*num_samples], condition)
            samples.append(sample)
            
        return torch.cat(samples).detach().numpy()
```

```python
def train_inn(model, train_data, train_labels, epochs, lr=0.001, batch_size=128, k=None):
   
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    n_classes = model.condition_size
    for epoch in range(epochs):
        total_loss = 0
        for i in range(0, len(train_data), batch_size):
            batch = torch.FloatTensor(train_data[i:i+batch_size])
            labels_batch = train_labels[i:i + batch_size]
            condition = torch.FloatTensor(np.eye(n_classes)[labels_batch]) 

            optimizer.zero_grad()

            z, log_det = model(batch, condition)
            nll_loss = 0.5 * torch.sum(z2, dim=1) - log_det
            nll_loss = torch.mean(nll_loss)

            if k is not None:  # Bottleneck
                z_recon = z.clone()
                z_recon[:, k:] = 0
                x_recon = model.inverse(z_recon, condition) 
                recon_loss = torch.mean((batch - x_recon)2)
                loss = nll_loss + recon_loss
            else:
                loss = nll_loss

            loss.backward()
            optimizer.step()

            total_loss += loss.item()

        if (epoch + 1) % 10 == 0:
            loss_str = f'NLL Loss: {nll_loss.item():.4f}'
            if k is not None:
                loss_str = f'Combined Loss: {loss.item():.4f}, NLL Loss: {nll_loss.item():.4f}, Recon Loss: {recon_loss.item():.4f}'
            print(f'Epoch {epoch+1}, {loss_str}')
```

```python
def main():
    input_size = 64
    hidden_size = 256 
    blocks = 8       
    n_samples = 200
    epochs = 100
    k_bottleneck = 24  

    
    X_digits, y_digits = load_digits(return_X_y=True)
    X_digits = X_digits / 16.0 - 0.5 
    n_classes_digits = len(np.unique(y_digits))

    model_digits_conditional = RealNVP(input_size, hidden_size, blocks, n_classes_digits)
    train_inn(model_digits_conditional, X_digits, y_digits, epochs) # Pass labels




    # --- MNIST Dataset ---
    transform = transforms.Compose([
        DownsampleTransform(target_shape=(8,8)),
        transforms.Grayscale(num_output_channels=1),
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,)) 
    ])


    mnist_dataset = datasets.MNIST(root='./data', train=True, transform=transform, download=True)
    X_mnist, y_mnist = np.array([x.numpy().reshape(input_size) for x, _ in mnist_dataset])
    n_classes_mnist = len(np.unique(y_mnist))

    model_mnist_conditional = RealNVP(input_size, hidden_size, blocks, n_classes_mnist)
    train_inn(model_mnist_conditional, X_mnist, y_mnist, epochs) 

    def visualize_samples(model, X_test, y_test, title, n_classes):
        fig, axes = plt.subplots(n_classes, 10, figsize=(10, n_classes))
        for label in range(n_classes):
            condition = torch.FloatTensor(np.eye(n_classes)[label]).unsqueeze(0) 
            generated_samples = model.sample(10, condition.repeat(10, 1)) 


            for i in range(10):
                axes[label, i].imshow(generated_samples[i].reshape(8, 8), cmap='gray')
                axes[label, i].axis('off')


        plt.suptitle(f"{title} - Generated Samples (Conditional)")
        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
        plt.show()

      
        generated_all = []
        y_gen_all = []
        for label in range(n_classes):
            condition = torch.FloatTensor(np.eye(n_classes)[label]).unsqueeze(0)
            gen_samples = model.sample(1000, condition.repeat(1000, 1))  
            generated_all.append(gen_samples)
            y_gen_all.extend([label] * 1000)



        X_gen_all = np.concatenate(generated_all)
        y_gen_all = np.array(y_gen_all)


        clf = RandomForestClassifier(max_depth=2, random_state=0)
        clf.fit(X_test, y_test)


        accuracy = clf.score(X_gen_all, y_gen_all)
        print(f"{title} Random Forest Accuracy on generated samples: {accuracy}")

    visualize_samples(model_digits_conditional, X_digits, y_digits, "Digits (Conditional)", n_classes_digits)

    visualize_samples(model_mnist_conditional, X_mnist, y_mnist, "MNIST (Conditional)", n_classes_mnist)

if __name__ == "__main__":
    main()
```
